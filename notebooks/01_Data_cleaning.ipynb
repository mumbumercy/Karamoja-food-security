#defining the path for the second data (district crop yield)

```python
district_path = '../data/raw/DATA/TABLES/Uganda_Karamoja_District_Crop_Yield_Population.csv'
```

#loading the district crop yield data and confirming the code

```python
df_district = pd.read_csv('../data/raw/DATA/TABLES/Uganda_Karamoja_District_Crop_Yield_Population.csv')
print("District Data Loaded Successfully!")
df_district.head()
```

#cleaning the data,renaming the columns this will be used later to merge the data sets ready for analysis 

```python
df_district = df_district.rename(columns={
    'NAME': 'District_Name',          
    'POP': 'Population',
    'S_Yield_Ha': 'Sorghum_Yield_Ha',
    'M_Yield_Ha': 'Maize_Yield_Ha',
    'S_Area_Ha': 'Sorghum_Area_Ha',
    'M_Area_Ha': 'Maize_Area_Ha',
    'S_Prod_Tot': 'Sorghum_Prod_Tot',
    'M_Prod_Tot': 'Maize_Prod_Tot',
    'Crop_Area_Ha': 'Total_Crop_Area_Ha'
})
```

```python
df_district.head()
```

#making the column titles to have title case

```python
df_district.columns = df_district.columns.str.strip().str.replace(' ', '_').str.title()
df_district.head()
```

#checking brief description of our data with the column data types 

```python
df_district.info()
```

merging the two data sets 

```python
df_merged = pd.merge(df, df_district, on='District_Name', how='left', suffixes=('', '_District'))
```

```python
print(f"Merged Data Shape: {df_merged.shape}")
df_merged.head()
```

#calculating the total food production (maize & sorgum) in tonnes

```python
df_merged['Total_Food_Tonnes'] = (df_merged['Sorghum_Prod_Tot'] + df_merged['Maize_Prod_Tot']) / 1000
```

#calculating total food per person in kgs

```python
df_merged['Food_Per_Person_Kg'] = (df_merged['Sorghum_Prod_Tot'] + df_merged['Maize_Prod_Tot']) / df_merged['Population']
```

calculating the dominant food

```python
df_merged['Primary_Crop'] = df_merged.apply(
    lambda x: 'Sorghum' if x['Sorghum_Prod_Tot'] >= x['Maize_Prod_Tot'] else 'Maize',
    axis=1
)
```
