#loading data 

```python
df = pd.read_csv('../data/raw/DATA/TABLES/Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv')
```

#checking if our code above worked  by seeing a short descriptiion of our data

```python
df.head()
```

```python
df['SUBCOUNTY_NAME'] = df['SUBCOUNTY_NAME'].str.title().str.strip()
```

```python
df['DISTRICT_NAME'] = df['DISTRICT_NAME'].str.title().str.strip()
```

#standardizing the names for the district and subcounty to title case

```python
df.columns = [c.replace(' ', '_') for c in df.columns]
```

```python
df.columns
```

```python
df[['DISTRICT_NAME', 'SUBCOUNTY_NAME', 'S_Yield_Ha']].head()
```

#standardizing the names for the columns to title case

```python
df.columns = df.columns.str.strip().str.replace(' ', '_').str.title()
```

```python
df.columns
```

checking if our code above worked

```python
df.head(3)
```

#making the column titles more readable

```python
df = df.rename(columns={
    'Pop': 'Population',
    'S_Yield_Ha': 'Sorghum_Yield_Ha',
    'M_Yield_Ha': 'Maize_Yield_Ha',
    'S_Area_Ha': 'Sorghum_Area_Ha',
    'M_Area_Ha': 'Maize_Area_Ha'
})
```

```python
df.head(3)
```

#finding the missing values and replacing them

```python
missing_counts = df.isnull().sum()
```

```python
cols_to_fill = ['Sorghum_Yield_Ha', 'Maize_Yield_Ha', 'Sorghum_Area_Ha', 'Maize_Area_Ha']
df[cols_to_fill] = df[cols_to_fill].fillna(0)
```
